```python
# Import required modules
from transformers import pipeline, set_seed

# Set up text generation pipeline using GPT-2
generator = pipeline('text-generation', model='gpt2')
set_seed(42)
```

```python
# Generate text for a given prompt
prompt = "The future of artificial intelligence in healthcare"
output = generator(prompt, max_length=500, num_return_sequences=5)

# Display the generated text
print("Generated Text (GPT-2):\n", output[0]['generated_text'])
```
